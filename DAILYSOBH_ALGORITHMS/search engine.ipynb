```python
import pandas as pd

# Load the CSV file
file_path = 'daily_sobh_news.csv'
data = pd.read_csv(file_path)

# Define the search function
def search_news(data, query):
    # Filter the data based on the given query (case insensitive)
    filtered_data = data[data['Topic'].str.contains(query, case=False, na=False) |
                         data['Title'].str.contains(query, case=False, na=False)]
    return filtered_data

# Example interactive user input (for demonstration purposes)
def user_input_search():
    query = input("Enter your search query: ")
    results = search_news(data, query)
    if results.empty:
        print("No results found for your query.")
    else:
        print(results)

# Call the user input search function
user_input_search()
```

```python
import pandas as pd

# Load the CSV file
file_path = 'daily_sobh_news.csv'
data = pd.read_csv(file_path)

# Extract unique topics from the data
unique_topics = data['Topic'].unique()

# Define the search function
def search_news_by_topic(data, topic):
    # Filter the data based on the given topic (case insensitive)
    filtered_data = data[data['Topic'].str.contains(topic, case=False, na=False)]
    return filtered_data

# Function to display the menu and get user input
def user_input_search():
    print("Available topics:")
    for idx, topic in enumerate(unique_topics, start=1):
        print(f"{idx}. {topic}")
    
    # Ensure valid input from user
    while True:
        try:
            choice = int(input("Enter the number corresponding to your search topic: "))
            if 1 <= choice <= len(unique_topics):
                selected_topic = unique_topics[choice - 1]
                break
            else:
                print("Invalid choice. Please enter a number from the list.")
        except ValueError:
            print("Invalid input. Please enter a number.")

    # Perform search and display results
    results = search_news_by_topic(data, selected_topic)
    if results.empty:
        print("No results found for your topic.")
    else:
        print(results)

# Call the user input search function
user_input_search()
```
